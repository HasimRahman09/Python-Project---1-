##Take appropriate actions to handle the outliers.


```python
#Example from data above
# Define a function to remove outliers based on the IQR(Interquartile range) method for specific columns (we will not be removing outliers from the salary columns)
def remove_outliers_iqr(column):
    if column.name in ['Rating', 'Founded', 'employer_provided', 'age']:
        Q1 = column.quantile(0.25)
        Q3 = column.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        return column[(column >= lower_bound) & (column <= upper_bound)]
    else:
        return column  # Return other columns as they are

# Create a new DataFrame to store the cleaned data
df_cleaned = df.copy()

# Apply the remove_outliers_iqr function to specific columns
df_cleaned = df_cleaned.apply(remove_outliers_iqr)

# Display the cleaned DataFrame
print("Original DataFrame:")
print(df[['Rating', 'Founded', 'employer_provided', 'age']])
print("\nDataFrame with Outliers Removed:")
print(df_cleaned[['Rating', 'Founded', 'employer_provided', 'age']])


# Apply the remove_outliers_iqr function to each column
df_cleaned = df.apply(remove_outliers_iqr)

# Display the cleaned DataFrame
print("Original DataFrame:")
print(df)
print("\nDataFrame with Outliers Removed:")
print(df_cleaned)
```

```python
# Showing the removed outliers after removing the box plots
columns_to_plot = ['Rating', 'Founded','employer_provided','age']

# Create a box plot for each column
fig, axes = plt.subplots(nrows=1, ncols=len(columns_to_plot), figsize=(12, 6))
fig.suptitle('Box Plots for the columns with outliers removed', fontsize=20)

for i, column in enumerate(columns_to_plot):
    sns.boxplot(x=df_cleaned[column], ax=axes[i])
    axes[i].set_title(column)

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.9])
plt.show()
```

#Which Programming language leads to a higher salary?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter the data to include only the rows with 'python_yn', 'R_yn', 'spark', 'aws', or 'excel'
filtered_data = df[(df['python_yn'] == 1) | (df['R_yn'] == 1) | (df['spark'] == 1) | (df['aws'] == 1) | (df['excel'] == 1)]

# Set the figure size
plt.figure(figsize=(12, 6))

# Create separate box plots for each programming language
plt.subplot(1, 5, 1)
sns.boxplot(x='python_yn', y='avg_salary', data=filtered_data, color='skyblue')
plt.title('Python')

plt.subplot(1, 5, 2)
sns.boxplot(x='R_yn', y='avg_salary', data=filtered_data, color='skyblue')
plt.title('R')

plt.subplot(1, 5, 3)
sns.boxplot(x='spark', y='avg_salary', data=filtered_data, color='skyblue')
plt.title('Spark')

plt.subplot(1, 5, 4)
sns.boxplot(x='aws', y='avg_salary', data=filtered_data, color='skyblue')
plt.title('AWS')

plt.subplot(1, 5, 5)
sns.boxplot(x='excel', y='avg_salary', data=filtered_data, color='skyblue')
plt.title('Excel')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()


```

# Correlation Heatmap:

Create a correlation heatmap to visualize the pairwise correlation between numerical columns. This can help identify relationships between columns.

Finding the correlation for all the numerical variables?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Select the columns of interest
columns_of_Heatmap = ['Rating', 'employer_provided', 'min_salary', 'max_salary', 'avg_salary', 'age']

# Create a DataFrame with only the selected columns
selected_data = df[columns_of_Heatmap]

# Compute the correlation matrix
correlation_matrix = selected_data.corr()

# Create a heatmap with the "Pastel1" colormap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Pastel1', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
